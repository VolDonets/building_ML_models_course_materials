# Theory: How to use backpropagation on a single perceptron?

## I. Equation for `binary cross-entropy`:

$$
Loss = - \frac{1}{N} \sum_{i=1}^{N} [y_i * log(y^*_i) + (1-y_i) * log(1 - y^*_i)]
$$
, where:
 
* $ N $ - number of samples
* $ y_i $ - probability of correct class (in case of binary problem 0 for 0-th class and 1 for 1-st class)
* $ y^*_i $ - collected value of class probability (by your neural network or any other kind of NN)

## II. Equation for a single perceptron:

For a single perceptron with two inputs ( $x_1$ and $x_2$), the forward pass consist of two steps: the linear sum and the activation function

1. Linear sum:

$$
z = x_1 * w_1 + x_2 * w_2 + b
$$

2. Sigmoid activation function:

$$
y_{pred} = \sigma(z) = \frac{1}{1 + exp(-z)}
$$


## III. Loss function for a single perceptron:

For a single example $X=(x_1, x_2)$, $y_{true}$ the `binary cross-entropy` will look like this: 

$$
Loss = -[y_{true} * log(y_{pred}) + (1 - y_{true}) * log(1 - y_{pred})]
$$

The idea of backpropagation is to use the chain rule of calculus to find the gradient, which tells us the influence of each weight on the final loss.

And for each weight adjustment temp will look like this:

$$
\frac{\delta L}{\delta w_1} = \frac{\delta L}{\delta y_{pred}} * \frac{\delta y_{pred}}{\delta z} * \frac{\delta z}{\delta w_1}
$$

where $z$ = $x_1 * w_1 + x_2 * w_2 + b$  -- linear sum.

Let's break down the partial derivatives:

1. Derivative of the loss function with respect to the prediction:

$$
\frac{\delta L}{\delta y_{pred}} = \frac{y_{pred} - y_{true}}{y_{pred}(1 - y_{pred})}
$$

2. Derivative of the sigmoid activation with respect to the linear sum:

$$
\frac{\delta y_{pred}}{\delta z} = y_{pred} * (1 - y_{pred})
$$

3. Derivative of the linear sum with respect to the weights:

$$
\frac{\delta z}{\delta w_1} = x_1
$$


Calculating the final gradients:

When we multiply these together, notice how perfectly the denominator of the loss derivative cancels out the sigmoid derivative! This is the mathematical advantage of pairing sigmoid with binary cross-entropy.

$$
\triangle w_1 = \frac{\delta L}{\delta w_1} = \frac{y_{pred} - y_{true}}{y_{pred}(1 - y_{pred})} * y_{pred} * (1 - y_{pred}) * x_1 = [y_{pred} - y_{true}] * x_1
$$

$$
\triangle w_2 = \frac{\delta L}{\delta w_2} = \frac{y_{pred} - y_{true}}{y_{pred}(1 - y_{pred})} * y_{pred} * (1 - y_{pred}) * x_2 = [y_{pred} - y_{true}] * x_2
$$

$$
\triangle b = \frac{\delta L}{\delta b} = \frac{y_{pred} - y_{true}}{y_{pred}(1 - y_{pred})} * y_{pred} * (1 - y_{pred}) * 1 = = [y_{pred} - y_{true}] * 1
$$


So adjusted weights to minimize loss value would be:

$$
w_1^{new} = w_1 - a * \triangle w_1
$$

$$
w_2^{new} = w_2 - a * \triangle w_2
$$

$$
b^{new} = b - a * \triangle b
$$


In case of multiple training samples $X = {(x_{1, 1}, x_{1, 2}), (x_{2, 1}, x_{2, 2}), ..., (x_{n, 1}, x_{n, 2})}; y = {y_1, y_2, ..., y_n}$ calculate the gradients for each sample and average them (since our original loss function in Section I divides by $N$) before applying the final weight update.

## 1. Load the data

```python
# Load the data from file

dataset = pd.read_csv("./src/IRIS.csv")
x_data = dataset[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
```

```python
# Train / Test split the data

train_count = int(len(x_data) * 0.8)

x_train = x_data[:train_count]
y_train = y_data[:train_count]

x_test = x_data[train_count:]
y_test = y_data[train_count:]
```

## 2. Define the model's class

```python
# Define the ReLU activation function
def relu(x):
    return np.maximum(x, 0)


# Define the softmax activation function
def softmax(x):
    e_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return e_x / np.sum(e_x, axis=1, keepdims=True)


# Define a neural network with one hidden layer
class NeuralNetwork:
    def __init__(self, input_size, hidden_size_0, hidden_size_1, output_size):
        # ReLU 0
        self.W0 = np.random.randn(input_size, hidden_size_0) * 0.01
        self.b0 = np.zeros((1, hidden_size_0))

        # ReLU 1 / sigmoid
        self.W1 = np.random.randn(hidden_size_0, hidden_size_1) * 0.01
        self.b1 = np.zeros((1, hidden_size_1))

        # Softmax 2 / sigmoid
        self.W2 = np.random.randn(hidden_size_1, output_size) * 0.01
        self.b2 = np.zeros((1, output_size))

    def forward(self, X):
        # ReLU 0
        self.Z0 = np.dot(X, self.W0) + self.b0
        self.A0 = relu(self.Z0)

        # ReLU 1
        self.Z1 = np.dot(self.A0, self.W1) + self.b1
        self.A1 = relu(self.Z1)

        # Softmax 2
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        self.A2 = softmax(self.Z2)
        return self.A2

    def backward(self, X, y, learning_rate):
        num_examples = X.shape[0]

        # Softmax 2
        dZ2 = self.A2 - y  # dL / dF
        dW2 = np.dot(self.A1.T, dZ2) / num_examples
        db2 = np.sum(dZ2, axis=0, keepdims=True) / num_examples
        dA1 = np.dot(dZ2, self.W2.T)

        # ReLU 1
        dZ1 = dA1 * (self.Z1 > 0)
        dW1 = np.dot(self.A0.T, dZ1) / num_examples
        db1 = np.sum(dZ1, axis=0, keepdims=True) / num_examples
        dA0 = np.dot(dZ1, self.W1.T)
        
        # ReLU 0
        dZ0 = dA0 * (self.Z0 > 0)
        dW0 = np.dot(X.T, dZ0) / num_examples
        db0 = np.sum(dZ0, axis=0, keepdims=True) / num_examples

        # Training of the ANN
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1
        self.W0 -= learning_rate * dW0
        self.b0 -= learning_rate * db0
```

```python
def compute_categorical_cross_entropy_loss(y_true, y_pred):
    """
    Calculate categorical cross-entropy loss.

    Arguments:
    y_true -- true labels (one-hot encoded)
    y_pred -- predicted probability distribution over the classes (softmax output)

    Returns:
    loss -- categorical cross-entropy loss
    """
    # Add a small epsilon value to prevent division by zero
    epsilon = 1e-8

    # Calculate categorical cross-entropy loss
    N = y_true.shape[0]
    loss = -1/N * np.sum(y_true * np.log(y_pred + epsilon))

    return loss
```

```python
def compute_accuracy(y_true, y_pred):
    """
    Compute the accuracy of the predicted probabilities relative to the true labels.

    Arguments:
    y_true -- a numpy array of shape (num_examples, num_classes) containing the true labels in one-hot format
    y_pred -- a numpy array of shape (num_examples, num_classes) containing the predicted probabilities

    Returns:
    accuracy -- a scalar representing the fraction of examples that were classified correctly
    """
    # Determine the predicted classes
    y_pred_classes = np.argmax(y_pred, axis=1)

    # Determine the true classes
    y_true_classes = np.argmax(y_true, axis=1)

    # Compute the accuracy
    num_correct = np.sum(y_pred_classes == y_true_classes)
    num_examples = y_true.shape[0]
    accuracy = num_correct / num_examples

    return accuracy
```

## 3. Build (compile) and train the ANNs model

```python
# Create a neural network
# I've made very weird ANN model with 1 neuron (narrow bottleneck) on the beginning, 
# and 25 neurons further.
# This was done to make training process on an relatively simple data more appealing.
nn = NeuralNetwork(
    input_size=4, 
    hidden_size_0=1, 
    hidden_size_1=25, 
    output_size=3,
)

# Train the neural network
learning_rate = 0.1
num_epochs = 5_000

val_loss_history = []
val_acc_history = []

train_loss_history = []
train_acc_history = []

for i in tqdm(range(num_epochs)):
    # Perform a forward pass through the network
    y_pred_train = nn.forward(x_train)

    # Compute the loss and accuracy on the training set
    loss_train = compute_categorical_cross_entropy_loss(y_train, y_pred_train)
    acc_train = compute_accuracy(y_train, y_pred_train)

    # Perform a backward pass through the network
    nn.backward(x_train, y_train, learning_rate)

    # Perform a forward pass through the network on the test set
    y_pred_test = nn.forward(x_test)

    # Compute the loss and accuracy on the test set
    loss_test = compute_categorical_cross_entropy_loss(y_test, y_pred_test)
    acc_test = compute_accuracy(y_test, y_pred_test)

    train_loss_history.append(loss_train)
    train_acc_history.append(acc_train)
    val_loss_history.append(loss_test)
    val_acc_history.append(acc_test)

    # Print the loss and accuracy on the training and test sets
    # if i % 100 == 0:
    #     print(f"Epoch {i}: train_loss={loss_train:.4f}, train_acc={acc_train:.4f}, test_loss={loss_test:.4f}, test_acc={acc_test:.4f}")
```

# II. Using library for implementation your models (Tensorflow & Keras)

```python
initializer = tf.keras.initializers.RandomNormal(mean=0.0, stddev=0.01)
```

```python
# Build the network architecture
nn = tf.keras.Sequential([
    # Layer 0: Hidden size = 1, ReLU
    tf.keras.layers.Input(shape=(4,)),
    tf.keras.layers.Dense(
        units=1, 
        activation='relu',
        kernel_initializer=initializer,
        bias_initializer='zeros',
        name="Hidden_Layer_0"
    ),
    
    # Layer 1: Hidden size = 25, ReLU
    tf.keras.layers.Dense(
        units=25, 
        activation='relu',
        kernel_initializer=initializer,
        bias_initializer='zeros',
        name="Hidden_Layer_1"
    ),
    
    # Layer 2: Output size = 3, Softmax
    tf.keras.layers.Dense(
        units=3, 
        activation='softmax',
        kernel_initializer=initializer,
        bias_initializer='zeros',
        name="Output_Layer"
    )
])
```

```python
# Define the optimizer and loss function to match your backward() pass
# SGD (Stochastic Gradient Descent) matches your weight update logic: W -= learning_rate * dW
learning_rate = 0.1 
optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate)

# CategoricalCrossentropy matches the derivative dZ2 = A2 - y 
loss_fn = tf.keras.losses.CategoricalCrossentropy()

# Compile the model
nn.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
```

```python
# Print a summary of the network parameters
nn.summary()
```

```python
epochs = 5000
batch_size = 25

history = nn.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epochs, batch_size=batch_size)
```

```python
# 3. Visualize the training history
# Extract the metrics from the history dictionary
loss = history.history['loss']
val_loss = history.history['val_loss']
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
epochs_range = range(1, epochs + 1)

# Create a figure with two subplots (1 row, 2 columns)
plt.figure(figsize=(12, 5))

# Plot 1: Training and Validation Loss
plt.subplot(1, 2, 1)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss',)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot 2: Training and Validation Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs_range, accuracy, label='Training Accuracy',)
plt.plot(epochs_range, val_accuracy, label='Validation Accuracy',)
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```
